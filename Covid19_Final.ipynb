```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv')
```

```python
covid_df
```

```python
covid_df.loc['Total']= covid_df.sum()
for i in range(0,11):
    covid_df.iloc[[3340], [i]] = "None" 
covid_df
```

```python
#This creates the dates list
dates = list(covid_df.columns.values)
dates = dates[11:]
print(dates)
```

```python
#This creates a list of the US cases
#Convert items in list to integer, do so for fairfield county and champaign county
US_cases = []
for i in range(11,len(covid_df.columns)):
    US_cases.append(covid_df.iloc[3340][i])

#print(US_cases)
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo
pyo.offline.init_notebook_mode(connected=True)
fig = go.Figure()
fig.add_trace(go.Scatter(x=dates, y=US_cases,
                    line_color='#FF0000',
                    mode='markers',
                    name='lines'
                        ))
fig.update_layout(
    title="Covid 19 Cases in the U.S",
    xaxis_title="Date",
    yaxis_title="Case Count",
)

value = US_cases[-1]
print("There are",f"{value:,}","confirmed cases in the US")
fig.show()
```

```python
uid = 84009001
row_num = covid_df[covid_df['UID'] == uid].index[0]

Fairfield_cases = []
for i in range(11,len(covid_df.columns)):
    Fairfield_cases.append(covid_df.iloc[row_num][i])
#print(Fairfield_cases)
```

```python
pyo.offline.init_notebook_mode(connected=True)
fig = go.Figure()
fig.add_trace(go.Scatter(x=dates, y=Fairfield_cases,
                    line_color='#0000FF',
                    mode='markers',
                    name='lines'
                        ))
fig.update_layout(
    title="Covid 19 Cases in the Fairfield County",
    xaxis_title="Date",
    yaxis_title="Case Count",
)

value = Fairfield_cases[-1]
print("There are",f"{value:,}","confirmed cases in the Fairfield County")
fig.show()
```

```python
uid = 84017019
row_num = covid_df[covid_df['UID'] == uid].index[0]

Champaign_cases = []
for i in range(11,len(covid_df.columns)):
    Champaign_cases.append(covid_df.iloc[row_num][i])
#print(Champaign_cases)
```

```python
pyo.offline.init_notebook_mode(connected=True)
fig = go.Figure()
fig.add_trace(go.Scatter(x=dates, y=Champaign_cases,
                    line_color='#008000',
                    mode='markers',
                    name='lines'
                        ))
fig.update_layout(
    title="Covid 19 Cases in the Champaign County",
    xaxis_title="Date",
    yaxis_title="Case Count",
)

value = Champaign_cases[-1]
print("There are",f"{value:,}","confirmed cases in the Champaign County")
fig.show()
```

```python
change = []
for i in range(1,len(Champaign_cases)):
    a = Champaign_cases[i] - Champaign_cases[i-1]
    change.append(a)
print(change)
```

```python
pyo.offline.init_notebook_mode(connected=True)
fig = go.Figure()
fig.add_trace(go.Scatter(x=dates, y=change,
                    line_color='#008000',
                    mode='markers',
                    name='lines'
                        ))
fig.update_layout(
    title="Increase in Covid 19 Cases in the Champaign County",
    xaxis_title="Date",
    yaxis_title="Case Count",
)
```
